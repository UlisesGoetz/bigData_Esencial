```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
from sklearn.cluster import KMeans
df= pd.read_csv("C:/Users/Mr.R/Downloads/archivos_base_python_data_science_big_data_esencial/base_datos_2008.csv",nrows=1000000)
newdf =df[["AirTime","DepDelay"]].dropna()
```

```python
newdf=df[["AirTime","Distance","TaxiOut","ArrDelay","DepDelay"]].dropna()
```

```python
kmeans = KMeans(n_clusters=7,random_state=0,n_jobs=-1).fit(newdf)
kmeans.labels_
```

```python
np.unique(kmeans.labels_,return_counts=True)
```

```python
plt.scatter(newdf["AirTime"],newdf["DepDelay"],c=kmeans.labels_)
plt.show()
```
